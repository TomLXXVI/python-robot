# Forward Kinematics
---

```python
import numpy as np
from python_robot.models import Planar3R
from python_robot.base import Frame

PI = np.pi
```

## Instantiation of the Robot - Initial Joint Configuration

The simple planar 3R robot arm will be used in this demo.

```python
robot = Planar3R(
    l1=1.0, l2=1.0,
    q_lim=((0.0, np.deg2rad(150.0)), (-PI, PI), (-PI, PI)),
)
```

```python
await robot.plot_async(extent=2.0, spacing=0.5, scale=0.2)
```

## Change the Joint Configuration of the Robot

Joint angles:

```python
theta1 = np.deg2rad(45)
theta2 = np.deg2rad(-30)
theta3 = np.deg2rad(0)
```

Determine the pose of the end-effector frame w.r.t. the base frame of the robot = the world frame:

```python
ee_frame = robot.fwd_kin(joint_coords=[theta1, theta2, theta3])
print(ee_frame.matrix)
```

> **NOTE**<br>
> Method `fwd_kin()` with given `joint_coords` does not change the internal configuration state of the robot. See image below: the robot is still in its initial configuration state. 

```python
await robot.plot_async(extent=2.0, spacing=0.5, scale=0.2)
```

To change the internal configuration state of the robot, attribute `joint_coords` of the robot must be set.

```python
robot.joint_coords = [theta1, theta2, theta3]
```

```python
await robot.plot_async(extent=2.0, spacing=0.5, scale=0.2)
```

Now call `fwd_kin()` without passing any joint coordinates (leave parameter `joint_coords` to default `None`):

```python
ee_frame = robot.fwd_kin()
print(ee_frame.matrix)
```

## Base Frame and Tool Frame

So far, the base frame of the robot coincided with the world frame, and the tool frame coincided with the last link frame of the robot. The base frame and the tool frame of the robot can be set at instantiation of the robot, but still also after instantiation, through attributes `base_frame` and `tool_frame` of the robot. The base frame is relative to the world frame, while the tool frame is relative to the last link frame of the robot.

```python
base_frame = Frame(origin=(0.2, 0.2, 0.0), orient_angles=(0, 0, 0))
tool_frame = Frame(origin=(0.1, 0.0, 0.0), orient_angles=(0, 0, -PI/4))
```

```python
robot.base_frame = base_frame
```

```python
await robot.plot_async(extent=2.0, spacing=0.5, scale=0.2)
```

```python
robot.tool_frame = tool_frame
```

```python
await robot.plot_async(extent=2.0, spacing=0.5, scale=0.2)
```

When we apply the same forward kinematics again, we observe that the base and tool frame have now also been taken into account. The base frame no longer coincides with the world frame (i.e. the station frame), but `ee_frame` still represents the pose of the end-effector frame (i.e. the tool frame) w.r.t. world frame. 

```python
ee_frame = robot.fwd_kin()
print(ee_frame.matrix)
```
